```python
import os
import numpy as np

from tflite_model_maker import model_spec
from tflite_model_maker import image_classifier
from tflite_model_maker.config import ExportFormat
from tflite_model_maker.config import QuantizationConfig
from tflite_model_maker.image_classifier import DataLoader

import matplotlib.pyplot as plt

```

```python
# set variables
base_path = '/content/gdrive/MyDrive/food-not-food/'
imagezipped = 'dataset.zip'
epochs = 20
```

```python
# unzipped images into session storage
# os.system("unzip -o {base_path}/{imagezipped}  -d .".format(base_path=base_path,imagezipped=imagezipped))
```

Create a dataloader

```python
# update carpet's path
train_dir = "/content/drive/MyDrive/food-not-food/training"
test_dir = "/content/drive/MyDrive/food-not-food/evaluation"
val_dir = "/content/drive/MyDrive/food-not-food/validation"
class_names = sorted(os.listdir(train_dir))
class_names
```

```python
train_data = DataLoader.from_folder(train_dir)
test_data = DataLoader.from_folder(test_dir)
val_data = DataLoader.from_folder(val_dir)
train_data, test_data

```

### Create model


```python
# Create model
model = image_classifier.create(train_data, 
                                model_spec=model_spec.get('efficientnet_lite0'),
                                validation_data = val_data,
                                epochs=epochs)

```

```python
# Evaluate the model
loss, accuracy = model.evaluate(test_data)
```

### Save model

```python
model.export(export_dir='.', tflite_filename="food_not_food_model_v3.tflite")
```
